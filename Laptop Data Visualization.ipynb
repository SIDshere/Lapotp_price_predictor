## Exploratory Data Analysis


```python

df.drop('Unnamed: 0',axis=1,inplace=True)
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Ram',data=df)

plt.xlabel('RAM') 
plt.ylabel('Number of laptops') 

plt.show()
     
```

## mean for different ram size

```python
df.groupby('Ram')['Price'].mean()
```

```python
sns.boxplot(x='Ram',y="Price",data=df)
plt.show()
```

```python
df['p_brand'].value_counts().plot(kind='bar')

plt.xlabel('Processor Name')
plt.ylabel('Number of laptops')
plt.show()
     
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10, 6))

# Create barplot
sns.barplot(x="Brand", y="Price", data=df)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
pd.pivot_table(df, index = ['Brand','processor_type'], values = 'Price').sort_values('Brand', ascending = False)
```

```python
# Set figure size
plt.figure(figsize=(10, 6))
sns.catplot(x="Brand", y="Price", hue="processor_type", data=df,height=6, aspect=10/6)
# Rotate x-axis labels
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
