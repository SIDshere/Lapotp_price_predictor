```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pages = list(range(1, 40))
data = []

for page in pages:
    req = requests.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(page)).text
    soup = BeautifulSoup(req, 'html.parser')

    laptops = soup.find_all('div', class_='_2kHMtA')
    for laptop in laptops:
        name_elem = laptop.find('div', class_='_4rR01T')
        name = name_elem.text.strip() if name_elem else None

        specs = laptop.find_all('li', class_='rgWa7D')
        specifications = {
            'Processor': None,
            'RAM': None,
            'Operating System': None,
            'Storage': None,
            'Display Size': None,
            'Warranty': None
        }
        for spec in specs:
            text = spec.text.strip()
            if "Core" in text:
                specifications['Processor'] = text
            elif "RAM" in text:
                specifications['RAM'] = text
            elif "HDD" in text or "SSD" in text:
                specifications['Storage'] = text
            elif "Operating" in text:
                specifications['Operating System'] = text
            elif "Display" in text:
                specifications['Display Size'] = text
            elif "Warranty" in text:
                specifications['Warranty'] = text

        price_elem = laptop.find('div', class_='_30jeq3 _1_WHN1')
        price = price_elem.text.strip() if price_elem else None

        rating_elem = laptop.find('div', class_='_3LWZlK')
        rating = rating_elem.text.strip() if rating_elem else None

        laptop_data = {
            'Name': name,
            'Processor': specifications['Processor'],
            'RAM': specifications['RAM'],
            'Operating System': specifications['Operating System'],
            'Storage': specifications['Storage'],
            'Display Size': specifications['Display Size'],
            'Warranty': specifications['Warranty'],
            'Price': price,
            'Rating': rating
        }

        data.append(laptop_data)

df2 = pd.DataFrame(data)

print(df2)
```

```python
def fill_null_with_brand_mode(df, brand_column, rating_column):
    unique_brands = df[brand_column].unique()
    for brand in unique_brands:
        brand_ratings = df.loc[df[brand_column] == brand, rating_column]
        brand_mode = brand_ratings.mode().values
        if len(brand_mode) > 0:
            brand_mode_value = brand_mode[0]
            df.loc[(df[brand_column] == brand) & (df[rating_column].isnull()), rating_column] = brand_mode_value
```

```python
def extract_ssd(value):
    if isinstance(value, str) and 'SSD' in value:
        size = value.split()[0]
        if size == '1':
            return '1000'
        elif size == '2':
            return '2000'
        elif size == '4':
            return '4000'
        else:
            return size
    else:
        return '0'
```
